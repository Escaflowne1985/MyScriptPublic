```python
import json
import pandas as pd
from pandas.io.json import json_normalize
import requests
import time
```

```python
with open("data.json",encoding="utf8") as f:
    data = json.load(f)
    
df_ = json_normalize(data, record_path=['data','category','urls'],
               meta=[['data','name'],['data','name','name']],
              errors='ignore')
```

```python
df_[df_["data.name"] == "内容分享"]
```

```python
df_.to_excel("data.xlsx")
```

```python
df = df_.dropna()
df.reset_index(drop=True,inplace=True)
```

```python
for i in range(len(df)):
    try:
        url = df["favicon"][i]
        title = df["title"][i]
        html = requests.get(url)
        print(url)
        with open("jpg/{}.jpg".format(title),'wb') as f:
            f.write(html.content)
    except:
        pass
```
