# Heatmap_Heatmap_with_label_show
- 【Python数据热力图AI教程】带有类别显示热力图
- 类别显示热力图

```python
import random

from pyecharts import options as opts
from pyecharts.charts import HeatMap
from pyecharts.faker import Faker

value = [[i, j, random.randint(0, 50)] for i in range(24) for j in range(7)]
c = (
    HeatMap()
    .add_xaxis(Faker.clock)
    .add_yaxis(
        "series0",
        Faker.week,
        value,
        label_opts=opts.LabelOpts(is_show=True, position="inside"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="HeatMap-Label Show"),
        visualmap_opts=opts.VisualMapOpts(),
    )
#     .render("heatmap_with_label_show.html")
)
c.render_notebook()
```

# Heatmap_Heatmap_base
- 【Python数据热力图AI教程】基础热力图
- 基础热力图

```python
import random

from pyecharts import options as opts
from pyecharts.charts import HeatMap
from pyecharts.faker import Faker

value = [[i, j, random.randint(0, 50)] for i in range(24) for j in range(7)]
c = (
    HeatMap()
    .add_xaxis(Faker.clock)
    .add_yaxis("series0", Faker.week, value)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="HeatMap-base"),
        visualmap_opts=opts.VisualMapOpts(),
    )
#     .render("heatmap_base.html")
)
c.render_notebook()
```
