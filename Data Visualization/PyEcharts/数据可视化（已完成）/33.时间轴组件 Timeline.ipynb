# Timeline_Timeline_pie
- 【Python数据可视化时间轴组件AI教程】时间轴&轴饼图
- 时间轴&轴饼图

```python
from pyecharts import options as opts
from pyecharts.charts import Pie, Timeline
from pyecharts.faker import Faker

attr = Faker.choose()
tl = Timeline()
for i in range(2015, 2020):
    pie = (
        Pie()
        .add(
            "Shop A",
            [list(z) for z in zip(attr, Faker.values())],
            rosetype="radius",
            radius=["30%", "55%"],
        )
        .set_global_opts(title_opts=opts.TitleOpts("year {} sales".format(i)))
    )
    tl.add(pie, "{}year".format(i))
# tl.render("timeline_pie.html")
tl.render_notebook()
```

# Timeline_Timeline_bar_with_graphic
- 【Python数据可视化时间轴组件AI教程】时间轴&柱状图
- 时间轴&柱状图

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Timeline
from pyecharts.commons.utils import JsCode
from pyecharts.faker import Faker

x = Faker.choose()
tl = Timeline()
for i in range(2015, 2020):
    bar = (
        Bar()
        .add_xaxis(x)
        .add_yaxis("Shop A", Faker.values())
        .add_yaxis("Shop B", Faker.values())
        .set_global_opts(
            title_opts=opts.TitleOpts("year {} sales - With Graphic".format(i)),
            graphic_opts=[
                opts.GraphicGroup(
                    graphic_item=opts.GraphicItem(
                        rotation=JsCode("Math.PI / 4"),
                        bounding="raw",
                        right=100,
                        bottom=110,
                        z=100,
                    ),
                    children=[
                        opts.GraphicRect(
                            graphic_item=opts.GraphicItem(
                                left="center", top="center", z=100
                            ),
                            graphic_shape_opts=opts.GraphicShapeOpts(
                                width=400, height=50
                            ),
                            graphic_basicstyle_opts=opts.GraphicBasicStyleOpts(
                                fill="rgba(0,0,0,0.3)"
                            ),
                        ),
                        opts.GraphicText(
                            graphic_item=opts.GraphicItem(
                                left="center", top="center", z=100
                            ),
                            graphic_textstyle_opts=opts.GraphicTextStyleOpts(
                                text="year {} sales".format(i),
                                font="bold 26px Microsoft YaHei",
                                graphic_basicstyle_opts=opts.GraphicBasicStyleOpts(
                                    fill="#fff"
                                ),
                            ),
                        ),
                    ],
                )
            ],
        )
    )
    tl.add(bar, "{}year".format(i))
# tl.render("timeline_bar_with_graphic.html")
tl.render_notebook()
```

# Timeline_Timeline_bmap
- 【Python数据可视化时间轴组件AI教程】时间轴&热力图
- 时间轴&热力图

```python
from pyecharts import options as opts
from pyecharts.charts import BMap, Timeline
from pyecharts.faker import Faker

tl = Timeline()
tl.add_schema(pos_left="50%", pos_right="10px", pos_bottom="15px")
for i in range(2015, 2020):
    bmap = (
        BMap()
        .add_schema(baidu_ak="FAKE_AK", center=[120.13066322374, 30.240018034923])
        .add(
            "bmap",
            [list(z) for z in zip(Faker.provinces, Faker.values())],
            type_="heatmap",
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Timeline-BMap-{}year".format(i)),
            visualmap_opts=opts.VisualMapOpts(pos_bottom="center", pos_right="10px"),
            tooltip_opts=opts.TooltipOpts(formatter=None),
        )
    )
    tl.add(bmap, "{}year".format(i))
# tl.render("timeline_bmap.html")
tl.render_notebook()
```

# Timeline_Timeline_sankey
- 【Python数据可视化时间轴组件AI教程】时间轴&桑葚图
- 时间轴&桑基图

```python
from pyecharts import options as opts
from pyecharts.charts import Sankey, Timeline
from pyecharts.faker import Faker

tl = Timeline()
names = ("Shop A", "Shop B", "Shop C")
nodes = [{"name": name} for name in names]
for i in range(2015, 2020):
    links = [
        {"source": names[0], "target": names[1], "value": Faker.values()[0]},
        {"source": names[1], "target": names[2], "value": Faker.values()[0]},
    ]
    sankey = (
        Sankey()
        .add(
            "sankey",
            nodes,
            links,
            linestyle_opt=opts.LineStyleOpts(opacity=0.2, curve=0.5, color="source"),
            label_opts=opts.LabelOpts(position="right"),
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="{} Shop（A, B, C）Sales".format(i))
        )
    )
    tl.add(sankey, "{}year".format(i))
tl.render("timeline_sankey.html")
tl.render_notebook()
```

# Timeline_Timeline_bar_reversal
- 【Python数据可视化时间轴组件AI教程】时间轴&横向柱状图
- 时间轴&横向柱状图

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Timeline
from pyecharts.faker import Faker

tl = Timeline()
for i in range(2015, 2020):
    bar = (
        Bar()
        .add_xaxis(Faker.choose())
        .add_yaxis("Shop A", Faker.values(), label_opts=opts.LabelOpts(position="right"))
        .add_yaxis("Shop B", Faker.values(), label_opts=opts.LabelOpts(position="right"))
        .reversal_axis()
        .set_global_opts(
            title_opts=opts.TitleOpts("Timeline-Bar-Reversal (Year: {} )".format(i))
        )
    )
    tl.add(bar, "{} year".format(i))
# tl.render("timeline_bar_reversal.html")
tl.render_notebook()
```

# Timeline_Timeline_map
