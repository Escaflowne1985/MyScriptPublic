# Gauge_Gauge
- 【Python数据可视化仪表盘AI教程】简单仪表盘
- 简单仪表盘

```python
import pyecharts.options as opts
from pyecharts.charts import Gauge



c = (
    Gauge(init_opts=opts.InitOpts())
    .add(series_name="Indicator", data_pair=[["completion", 15.5]])
    .set_global_opts(
        legend_opts=opts.LegendOpts(is_show=False),
        tooltip_opts=opts.TooltipOpts(is_show=True, formatter="{a} <br/>{b} : {c}%"),
    )
#     .render("gauge.html")
)
c.render_notebook()
```

# Gauge_Gauge_color
- 【Python数据可视化仪表盘AI教程】自定义颜色仪表盘
- 可以自定义区间刻度颜色的仪表盘

```python
from pyecharts import options as opts
from pyecharts.charts import Gauge

c = (
    Gauge()
    .add("Indicator",[("completion", 55.5)],
        axisline_opts=opts.AxisLineOpts(
            linestyle_opts=opts.LineStyleOpts(
                color=[(0.3, "#67e0e3"), (0.7, "#37a2da"), (1, "#fd666d")], width=30
            )
        ),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Gauge_color"),
        legend_opts=opts.LegendOpts(is_show=False),
    )
#     .render("gauge_color.html")
)
c.render_notebook()
```

# Gauge_Gauge_label_title_setting
- 【Python数据可视化仪表盘AI教程】定义字体的仪表盘
- 可以自定义仪表盘内文字的颜色和字体的仪表盘

```python
from pyecharts import options as opts
from pyecharts.charts import Gauge

c = (
    Gauge()
    .add(
        "",
        [("completion", 66.6)],
        title_label_opts=opts.LabelOpts(
            font_size=20, color="black", font_family="Microsoft YaHei"
        ),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Gauge_label_title_setting"))
#     .render("gauge_label_title_setting.html")
)
c.render_notebook()
```

# Gauge_Gauge_splitnum_label
- 【Python数据可视化仪表盘AI教程】分割段数的仪表盘
- 可以分割段数的仪表盘

```python
from pyecharts import options as opts
from pyecharts.charts import Gauge

c = (
    Gauge()
    .add(
        "Indicator",
        [("completion", 55.5)],
        split_number=5,
        axisline_opts=opts.AxisLineOpts(
            linestyle_opts=opts.LineStyleOpts(
                color=[(0.3, "#67e0e3"), (0.7, "#37a2da"), (1, "#fd666d")], width=30
            )
        ),
        detail_label_opts=opts.LabelOpts(formatter="{value}"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Gauge_splitnum_label"),
        legend_opts=opts.LegendOpts(is_show=False),
    )
#     .render("gauge_splitnum_label.html")
)
c.render_notebook()
```

# Gauge_change_radius
- 【Python数据可视化仪表盘AI教程】按比例缩放的仪表盘
- 可以根据比例修改的仪表盘

```python
from pyecharts import options as opts
from pyecharts.charts import Gauge

c = (
    Gauge()
    .add("", [("completion", 66.6)], radius="50%")
    .set_global_opts(title_opts=opts.TitleOpts(title="Gauge_change_radius=50%"))
#     .render("gauge_change_radius.html")
)
c.render_notebook()
```
