# Graph_Graph_base
- 【Python数据可视化关系图AI教程】基础关系图
- 基础关系图

```python
from pyecharts import options as opts
from pyecharts.charts import Graph

nodes = [
    {"name": "Node 1", "symbolSize": 10},
    {"name": "Node 2", "symbolSize": 20},
    {"name": "Node 3", "symbolSize": 30},
    {"name": "Node 4", "symbolSize": 40},
    {"name": "Node 5", "symbolSize": 50},
    {"name": "Node 6", "symbolSize": 40},
    {"name": "Node 7", "symbolSize": 30},
    {"name": "Node 8", "symbolSize": 20},
]
links = []
for i in nodes:
    for j in nodes:
        links.append({"source": i.get("name"), "target": j.get("name")})
g = (
    Graph()
    .add("", nodes, links, repulsion=10000)
    .set_global_opts(title_opts=opts.TitleOpts(title="Graph-Base"))
#     .render("graph_base.html")
)
g.render_notebook()
```

# Graph_Graph_with_edge_options
- 【Python数据可视化关系图AI教程】节点间标签自定义关系图
- 自定义关系节点间说明的关系图

```python
from pyecharts import options as opts
from pyecharts.charts import Graph

nodes_data = [
    opts.GraphNode(name="Node 1", symbol_size=10),
    opts.GraphNode(name="Node 2", symbol_size=20),
    opts.GraphNode(name="Node 3", symbol_size=30),
    opts.GraphNode(name="Node 4", symbol_size=40),
    opts.GraphNode(name="Node 5", symbol_size=50),
    opts.GraphNode(name="Node 6", symbol_size=60),
]
links_data = [
    opts.GraphLink(source="Node 1", target="Node 2", value=2),
    opts.GraphLink(source="Node 2", target="Node 3", value=3),
    opts.GraphLink(source="Node 3", target="Node 4", value=4),
    opts.GraphLink(source="Node 4", target="Node 5", value=5),
    opts.GraphLink(source="Node 5", target="Node 6", value=6),
    opts.GraphLink(source="Node 6", target="Node 1", value=7),
]
g = (
    Graph()
    .add(
        "",
        nodes_data,
        links_data,
        repulsion=4000,
        edge_label=opts.LabelOpts(
            is_show=True, position="middle", formatter="data {c} of {b}"
        ),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Graph-GraphNode-GraphLink-WithEdgeLabel")
    )
#     .render("graph_with_edge_options.html")
)
g.render_notebook()
```

# Graph_Graph_with_options
- 【Python数据可视化关系图AI教程】节点标签自定义关系图
- 自定义关系节点的关系图

```python
from pyecharts import options as opts
from pyecharts.charts import Graph

nodes = [
    opts.GraphNode(name="Node 1", symbol_size=10),
    opts.GraphNode(name="Node 2", symbol_size=20),
    opts.GraphNode(name="Node 3", symbol_size=30),
    opts.GraphNode(name="Node 4", symbol_size=40),
    opts.GraphNode(name="Node 5", symbol_size=50),
]
links = [
    opts.GraphLink(source="Node 1", target="Node 2"),
    opts.GraphLink(source="Node 2", target="Node 3"),
    opts.GraphLink(source="Node 3", target="Node 4"),
    opts.GraphLink(source="Node 4", target="Node 5"),
    opts.GraphLink(source="Node 5", target="Node 1"),
]
g = (
    Graph()
    .add("", nodes, links, repulsion=4000)
    .set_global_opts(title_opts=opts.TitleOpts(title="Graph-GraphNode-GraphLink"))
#     .render("graph_with_options.html")
)
g.render_notebook()
```

# Graph_Graph_weibo
- 【Python数据可视化关系图AI教程】微博关系图示例
- 微博关系图示例

```python
import json

from pyecharts import options as opts
from pyecharts.charts import Graph

with open("data/weibo.json", "r", encoding="utf-8") as f:
    j = json.load(f)
nodes, links, categories, cont, mid, userl = j

g = (
    Graph()
    .add(
        "",
        nodes,
        links,
        categories,
        repulsion=50,
        linestyle_opts=opts.LineStyleOpts(curve=0.2),
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(
        legend_opts=opts.LegendOpts(is_show=False),
        title_opts=opts.TitleOpts(title="Graph-微博转发关系图"),
    )
#     .render("graph_weibo.html")
)
g.render_notebook()
```

# Graph_Graph_les_miserables
- 【Python数据可视化关系图AI教程】和弦关系图
- 和弦关系图

```python
import json

from pyecharts import options as opts
from pyecharts.charts import Graph


with open("data/les-miserables.json", "r", encoding="utf-8") as f:
    j = json.load(f)
    nodes = j["nodes"]
    links = j["links"]
    categories = j["categories"]

g = (
    Graph(init_opts=opts.InitOpts(width="1000px", height="600px"))
    .add(
        "",
        nodes=nodes,
        links=links,
        categories=categories,
        layout="circular",
        is_rotate_label=True,
        linestyle_opts=opts.LineStyleOpts(color="source", curve=0.3),
        label_opts=opts.LabelOpts(position="right"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Graph-Les Miserables"),
        legend_opts=opts.LegendOpts(orient="vertical", pos_left="2%", pos_top="20%"),
    )
#     .render("graph_les_miserables.html")
)
g.render_notebook()
```
