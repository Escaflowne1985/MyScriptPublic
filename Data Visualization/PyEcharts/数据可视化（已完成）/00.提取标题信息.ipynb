```python
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.faker import Faker

type_list = ["战略轰炸机","攻击/战斗机","特种作战飞机","预警/指挥机","空中加油机","运输机","直升机","教练机"]
value_1 = [158,2034,148,538,491,727,189,1187]
value_2 = [177,1166,0,124,15,571,1060,287]
value_3 = [120,1777,0,14,3,390,94,755]

c = (
    Bar()
    .add_xaxis(type_list)
    .add_yaxis("美国", value_1)
    .add_yaxis("俄罗斯", value_2)
    .add_yaxis("中国", value_3)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="美中俄空军实力对比", subtitle="仅通过抓取公开数据"),
        xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-15)),
        brush_opts=opts.BrushOpts(),
    )
#     .render("bar_with_brush.html")
)
c.render_notebook()
```
