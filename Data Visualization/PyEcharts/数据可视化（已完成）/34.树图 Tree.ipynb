# Tree_Tree_base
- 【Python数据可视化树图AI教程】基础树图
- 基础树图

```python
from pyecharts import options as opts
from pyecharts.charts import Tree


data = [
    {
        "children": [
            {"name": "B"},
            {
                "children": [{"children": [{"name": "I"}], "name": "E"}, {"name": "F"}],
                "name": "C",
            },
            {
                "children": [
                    {"children": [{"name": "J"}, {"name": "K"}], "name": "G"},
                    {"name": "H"},
                ],
                "name": "D",
            },
        ],
        "name": "A",
    }
]
c = (
    Tree()
    .add("", data)
    .set_global_opts(title_opts=opts.TitleOpts(title="Tree-base"))
#     .render("tree_base.html")
)
c.render_notebook()
```

# Tree_Tree_bottom_top
- 【Python数据可视化树图AI教程】下上方向树图
- 下上方向树图

```python
import json

from pyecharts import options as opts
from pyecharts.charts import Tree

with open("data/flare.json", "r", encoding="utf-8") as f:
    j = json.load(f)
c = (
    Tree()
    .add(
        "",
        [j],
        collapse_interval=2,
        orient="BT",
        label_opts=opts.LabelOpts(
            position="top",
            horizontal_align="right",
            vertical_align="middle",
            rotate=-90,
        ),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Tree-bottom_top"))
#     .render("tree_bottom_top.html")
)
c.render_notebook()
```
